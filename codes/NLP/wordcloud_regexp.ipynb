```python
from wordcloud import WordCloud
```

```python
# 한글 문장 리스트
sentences = [
    "너무나도 감동적인 영화였습니다. 마지막 장면은 정말로 잊을 수 없네요.",
    "배우들의 연기력이 돋보였던 작품. 특히 주연 배우의 연기가 인상적이었습니다.",
    "시나리오가 상당히 독특하네요. 예상치 못한 플롯 트위스트가 좋았습니다.",
    "음악과 영상미, 그리고 스토리까지 모든 것이 완벽한 영화입니다.",
    "조금 지루한 부분도 있었지만, 전체적으로 볼 만했습니다.",
    "주제가 굉장히 심오하네요. 한 번 보고 이해하기 어렵지만 여러 번 보면서 생각할 거리를 얻은 것 같아요.",
    "개인적으로는 코미디 요소가 좀 더 있었다면 좋았을 것 같아요.",
    "처음부터 끝까지 긴장감 넘치는 스토리로 몰입감 100%였습니다!",
    "생각보다 평점이 낮아서 기대 안 하고 봤는데, 정말 재밌게 본 영화 중 하나입니다!",
    "비록 상업적인 성공은 아니었지만 저에게 많은 감동과 공감을 준 작품입니다."
]
```

## 자연어 처리 위한 단어 생성

```python
# "맛있는 음식을 먹으러 갈까요?" -> 4개
# "여행 계획을 세우고 있는데 어디로 갈까요?" -> 6
# 10개이지만 동일한 단어는 겹칠 수 있다.
text = ' '.join(sentences)
text
```

```python
## 조사 빼기
# 는, 이, 은, 를, 을, 한, 가, ...
import re
pattern = r'[는은이가을를만의과에]'
text_regexp = re.sub(pattern=pattern, repl='', string=text) # params : 변경대상, 대상 대체, 어떤것을 
text_regexp
```

```python
from collections import Counter
text_regex_count = Counter(text_regexp.split()) # list를 count
```

```python
# 한글 폰트 경로 설정 (Windows 환경에 맞게 경로 수정)
font_path = 'C:/Windows/Fonts/malgun.ttf'

wordCloud = WordCloud(font_path=font_path, background_color='white')
wordCloud.generate_from_frequencies(text_regex_count)
```

```python
wordCloud.words_
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordCloud)
plt.axis('off')
plt.show()
```
