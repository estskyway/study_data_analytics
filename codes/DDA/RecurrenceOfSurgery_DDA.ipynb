### quest 2 (df_ROS_forpatient 기준)
- 직업군이 상위 2위에 속하는 항목 : value_count, DataFrame, iloc
- 체중 > 70, 신장 > 170, 흡연여부 없는분
- 해당 레코드 갯수 표현

```python
# 직업군이 상위 2위에 속하는 항목 : value_count, DataFrame, iloc
df_ROS_forpatient = df_ROS[['성별', '신장', '심혈관질환', '체중', '흡연여부', '연령', '직업', '혈액형', '재발여부']]
df_ROS_forpatient['직업'].value_counts()
```

```python
# 직업군이 상위 2위에 속하는 항목
# 체중 > 70, 신장 > 170, 흡연여부 없는분
condition = ((df_ROS_forpatient['직업'] == '사무직') | (df_ROS_forpatient['직업'] == '자영업')) &  (df_ROS_forpatient['체중'] > 70) & (df_ROS_forpatient['신장'] > 170) & (df_ROS_forpatient['흡연여부'] == 0) 
df_ROS_forpatient_cond = df_ROS_forpatient.loc[condition]
df_ROS_forpatient_cond
```
